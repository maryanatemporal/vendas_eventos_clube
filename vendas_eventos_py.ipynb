Iniciamos a análise importando a biblioteca do Python Pandas e o Numpy.
Em seguida carregamos a nossa base de dados em formato de csv


```
import pandas as pd
import numpy as np

df_vendas_eventos = pd.read_csv("vendas_eventos.csv", encoding='latin1', sep=";", low_memory = False)

#Visualização inicial
display(df_vendas_eventos.head(10))

#Verificando colunas
display(list(df_vendas_eventos.columns))

#Exibindo informações da tabela
display(df_vendas_eventos.info())

#Descrevendo informações como: média, mínimo, máximo
display(df_vendas_eventos.describe())

```

Tratamento de dados:










```
#Transformando para o formato de "pd.to_datetime": ano, mês e dia

df_vendas_eventos["data"]= pd.to_datetime(df_vendas_eventos["data"], format= "%d/%m/%Y")
display(df_vendas_eventos)
```

Criando uma nova coluna chamada "faturamento" para calcularmos o valor total do faturamento dos eventos

```
df_vendas_eventos["faturamento"] = df_vendas_eventos["ingressos_vendidos"] * df_vendas_eventos["valor_ingresso"]

display(df_vendas_eventos)
```

Nesta parte, iremos agrupar valores de colunas semelhantes, selecionamos as colunas utilizando o "groupby": evento. Em seguida, calculamos a soma total desses faturamento utilizando a função "sum()". Também ordenamos os valores dos faturamentos ( do maior para o menor).

```
faturamento_evento = df_vendas_eventos.groupby("evento")["faturamento"].sum()
faturamento_evento = faturamento_evento.sort_values(ascending=False)
display(faturamento_evento)
```
